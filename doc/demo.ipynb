This module shows how to generate Environmentally Extended Input-Output (EEIO) symmetric tables for Canada using Statistiques Canada economic data and various physical flows accounts as well as data from the NPRI.

```python
import openIO
import pandas as pd
import numpy as np
```

To create an object, you simply need to enter 2 parameters:
- the path to the STATCAN economic data for the year and classification level you chose (available here: https://www150.statcan.gc.ca/n1/en/catalogue/15-602-X)
- the classification to be used for the symmetric table: either "industry" or "product"


Users can enter an extra argument: the path to an IOT pxp folder of Exiobase. It will allow openIO to estimate international imports' emissions with Exiobase, rather than considering the same emmissions as local productions. This argument is optional, enter it if you want international imports to be estimated through Exiobase.

```python
tables = openIO.IOTables('C://Users/.../Detail level/',
                         'product',
                         'C://Users/.../IOT_2017_pxp')
```

You now have access to typical IOTs:
