The technology matrix (A) provides the normalized direct requirements to produce commodities/services (e.g., how much \\$ of electricity do I need to produce 1\$ of steel). It covers 13 Canadian provinces/territories, 44 countries and 5 Rest of the World regions.

```python
tables.A
```

The capital matrix (K), available if you selected endogenizing_capitals=True when creating the openIO-Canda system, represents the normalized capital requirements to produce commodities/services (e.g., how much \\$ of building do I need to produce 1$ of steel).

```python
tables.K
```

The value added matrix (R) represents the normalized added value required to produce commodities/services (e.g., how much \\$ of wages do I need to produce 1$ of steel)

```python
tables.R
```

Note that the sum of A + K + R should result in 1\\$ on each column

Also note that A and R have their non-normalized equivalents (Z & W), e.g., how much \\$ of XXX do I need to produce the enterity of the sectors output in the selected year.

The final demand matrix (Y) records the demand of final consumers: households, governments, non-profit institutions, international exports (and gross fixed capital formation if you did not endogenize capitals)

```python
tables.Y
```

The environmental flow matrix (F) records the total annual direct emissions of each sector.

```python
tables.F
```

The normalized equivalent of F is S, which thus records the emissions for the production of 1$ of commodity/service

```python
tables.S
```

Matrices F and S record the direct emissions of industries, but final consumers also emit emissions (e.g., using your car or heating your home with fuel). These emissions are recorded in the FY matrix

```python
tables.FY
```

All these emissions are translated into environmental impacts through the characterization matrix (C). This matrix is based on the characterization factors of the LCIA method IMPACT World+ v2.1

```python
tables.C
```

The technology matrix (A) record the direct requirements. To get the total requirements (essentially the life cycle approach aspect) you can look at the Leontief inverse matrix (L) **_[If your L matrix is empty then you need to run .calc()]_**

```python
tables.L
```

If you multiply the total requirements by the final demand to scale it up and then by the normalized direct emissions, you end up with the total life cycle emissions of each sector of the economy. This is provided by the E matrix.

```python
tables.E
```

Finally, if you multiply the total life cycle emissions by the characterization factors, you end up with the total life cycle impacts of each sector of the economy. This is provided by the D matrix.

```python
tables.D
```
